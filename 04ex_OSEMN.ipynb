**4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.**
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

```python
CC = []
with open('data/04EX/credit_card.dat', 'r') as file:
    for line in file:
        CC.append (line)

#lista in cui salta lo 0 
LIST = []
LL = []
L=''
CC_NUM=[]
for i in range(2,len(CC[0])-4, 6):
    LIST.append(i)
for i in range(0,len(LIST), 5):
    LL = LL + LIST [i:i+4]

for m in range(len(CC)-1):
    for i in LL:
        AA = str(int((CC[m][i:i+4]),2))
        L = L + AA
    CC_NUM.append(L)
    L=''
print (CC_NUM)
```

**5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.**

```python
import json
data = json.load(open('data/04EX/user_data.json'))
file_AE = open("data/04EX/user_data_AE.cvs", 'w')
for i in range (len (data)):
    if data[i]['CreditCardType'] == 'American Express':
        json.dump(data[i], file_AE)
        if i < len(data)-1:
            file_AE.write ('\n')
        
file_AE.close ()     
```
